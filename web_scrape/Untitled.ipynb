```python
import pandas as pd
import requests as r
```

```python
df = pd.read_json('flood.json')
```

```python
locations = [i['location'] for i in df['result']]
```

```python
locations
```

```python
from tqdm import tqdm_notebook
```

```python
results = []
for city_name in tqdm_notebook(locations):
    url = f'http://api.openweathermap.org/data/2.5/weather?appid=134666dd0e2cc872e1f9ae630e853b34&units=Imperial&q={city_name}'
    
    try:
        response = r.get(url).json()        
        results.append([response['coord']['lon'], response['coord']['lat']])
    except Exception as e:
#         print(e)
        next
```
